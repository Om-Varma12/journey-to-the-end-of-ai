```python
import os
from dotenv import load_dotenv

load_dotenv()

%load_ext autoreload
%autoreload 2
```

```python
from datetime import datetime

from langchain_groq import ChatGroq
from langchain.agents import create_agent
from langgraph.graph import MessagesState

from src.prompts import (
    PLANNER_PROMPT,
    FRONTEND_PROMPT,
    REVIEWER_PROMPT,
    SUPERVISOR_PROMPT,
)

from src.tools import write_file, read_file, list_files
from src.subagent_tool import create_subagent_tool
```

```python
model = ChatGroq(
    model="openai/gpt-oss-20b",
    temperature=0
)
```

```python
planner = {
    "name": "planner-agent",
    "description": "Designs the website architecture",
    "prompt": PLANNER_PROMPT,
    "tools": [write_file],
}

frontend = {
    "name": "frontend-agent",
    "description": "Builds React components",
    "prompt": FRONTEND_PROMPT,
    "tools": [write_file, read_file],
}

reviewer = {
    "name": "reviewer-agent",
    "description": "Reviews and improves React code",
    "prompt": REVIEWER_PROMPT,
    "tools": [read_file, write_file, list_files],
}

subagents = [planner, frontend, reviewer]
```

```python
task_tool = create_subagent_tool(
    subagents=subagents,
    model=model,
    state_schema=MessagesState
)
```

```python
agent = create_agent(
    model=model,
    tools=[task_tool],
    system_prompt=SUPERVISOR_PROMPT,
    state_schema=MessagesState
)
```

```python
result = agent.invoke(
    {
        "messages": [
            {
                "role": "user",
                "content": """
                    Create a simple React website for a personal portfolio.

                    The site should include:

                    - Home page
                    - Projects page
                    - Contact page
                """
            }
        ]
    }
)

print(result["messages"][-1].content)
```
